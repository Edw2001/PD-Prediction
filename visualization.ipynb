```python
import pandas as pd
import matplotlib.pyplot as plt
import os
```

```python
corpus = pd.read_excel("D:\\PaHaW\\PaHaW\\PaHaW\\PaHaW_files\\corpus_PaHaW.xlsx")
corpus
Health = corpus[corpus["Disease"]=="H"]
PD = corpus[corpus["Disease"]=="PD"]
if 75 in PD["ID"]:
    print("True")
else:
    print("False")
```

```python

path = "D:/PaHaW/PaHaW/PaHaW/PaHaW_public"
files = os.listdir(path)
for i in files:
    ID = int(i)
    if ID >= 10:
        file_name = "000"+str(ID)
    else:
        file_name = "0000"+str(ID)
    try:
        df = pd.read_csv("D:\\PaHaW\\PaHaW\\PaHaW\\PaHaW_public\\"+file_name+"\\"+file_name+"__1_1.svc", skiprows=1, sep=' ', header=None, names=['Y', 'X', 'Time', 'Button', 'Azimuth', 'Altitude', 'Pressure'])
    except:
        print("Missing spiral pattern:",file_name)
        continue
    plt.figure(figsize=(10, 6))

    plt.scatter(df[df['Button'] == 1]['X'], df[df['Button'] == 1]['Y'], 
                s=df[df['Button'] == 1]['Pressure']*0.1, color='blue', alpha=0.5, label='Pen Down')

    plt.scatter(df[df['Button'] == 0]['X'], df[df['Button'] == 0]['Y'], 
                s=df[df['Button'] == 0]['Pressure'], color='red', alpha=0.5, label='Pen Up')

    plt.title('Visualization of SVC Data with Pen Status and Pressure')
    plt.xlabel('X Coordinate')
    plt.ylabel('Y Coordinate')
    plt.legend()

    if ID in Health["ID"].tolist():
        plt.savefig("D:\\PaHaW\\PaHaW\\PaHaW\\PaHaW_visualization\\Health\\"+str(ID)+".png")
        print("Ploted Health Sample:",file_name)
    elif ID in PD["ID"].tolist():
        plt.savefig("D:\\PaHaW\\PaHaW\\PaHaW\\PaHaW_visualization\\PD\\"+str(ID)+".png")
        print("Ploted PD Sample:",file_name)
    else:
        print("Not in PD or Health:",file_name)
    plt.close()
```
